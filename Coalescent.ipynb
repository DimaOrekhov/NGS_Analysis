```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

matplotlib.rcParams.update({'font.size': 12}) 
```

```python
T = [0, 100,225,385,600,900,1350,2100,3600,8100] # adding 0
N_pop = len(T) #current number of populations
deltas = np.diff(T)
fig, ax = plt.subplots(figsize=[16, 9])
for p in T:
    ax.plot([p, p],[10, 20])
ax.plot(T, [15]*len(T), linestyle ='--')
ax.set_title('Convergence time dynamics')
```

## When $k \choose 2$ is large in present and decreases overtime 

```python
fig, ax = plt.subplots(figsize=[20, 9])
for i, d in enumerate(deltas):
    k = N_pop - i
    ax.plot(T[i:i+2], [0.5 * (k*(k-1)/2) * d] * 2, c='b')
ax.set_xlabel('$t$')
ax.set_ylabel('$N_e$')
ax.grid()
_ = ax.set_xticks(T)
```

## Oposite situation

```python
fig, ax = plt.subplots(figsize=[20, 9])
for i, d in enumerate(deltas):
    k = i
    ax.plot(T[i:i+2], [0.5 * (k*(k-1)/2) * d] * 2, c='b')
ax.set_xlabel('$t$')
ax.set_ylabel('$N_e$')
ax.grid()
_ = ax.set_xticks(T)
```
